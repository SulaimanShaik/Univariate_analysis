```python
def Graph_One(data):
    import pandas as pd                                  #Import package to do data manuplation
    import matplotlib.pyplot as plt                      #package for data visualisation
    data=pd.read_csv(data,index_col=0)                   #load csv file into data variable
    
    
    # For creating Histograms;
    for f in list(data.columns):                         #Select individual column iteratively
        data.hist(column=f,figsize=(8,6))                #plot histogram
        plt.savefig(str(f)+'_histogram.png')             #save plotted histograph in a directory
        plt.show()                                       #show graph
        
        #Plotting boxplot;
        data.boxplot(column=f,figsize=(8,6))             #plotting box plot
        plt.savefig(str(f)+'_boxplot.png')               #saving boxplot
        plt.show()
        
        
        
        
```

## 2.Univariate analysis for all the columns if no column is mentioned or mentioned column 

```python
def Graph_Two(df,columns=[]):
    import pandas as pd
    import matplotlib.pyplot as plt
    df=pd.read_csv(df,index_col=0)
    
    if len(columns)==0:                 #If no column is mentioned plot graphs for all the columns
        columns = list(df.columns)
        
    for f in columns:
        df.hist(column=f,grid=False,figsize=(8,6))
        plt.savefig(str(f)+'_hist.png')
        plt.show()
        
        df.boxplot(column=f,grid=False,figsize=(8,6))
        plt.savefig(str(f)+'_boxplot.png')
        plt.show()
```

```python
Graph_Two("cars.csv",["cylinders","class"])
```

## 3.Univariate Analysis of categorical and numerical data
### Categorical=barplot
### Numerical=boxplot and histograms

```python
def Graph_Three(data,columns=[]):
    import pandas as pd
    import numpy as np
    import seaborn as sb
    import matplotlib.pyplot as plt
    data=pd.read_csv(data,index_col=0)
    if len(columns)==0:
        columns=list(data.columns)
    num=[]
    cat=[]
    #Classifing numerical and categorical data into num[] and cat[] lists
    for i in columns:
        if((np.array(data[i].unique()).shape[0]>30 and data[i].dtype in['float64','int64'])):
            num.append(i)  #appending to num list
        else:
            cat.append(i)  #appending to cat list
           
           
    for j in num:
        data.hist(column=j,grid=False,figsize=(8,6))
        plt.savefig('%s_hist.png'%j)
           
           
        data.boxplot(column=j,grid=False,figsize=(8,6))
        plt.savefig('%s_boxplot.png'%j)
           
    for g in cat:
        sb.countplot(data[g])
        plt.savefig(str(g)+'_barplot.png')
```

```python
Graph_Three("cars.csv")
```

# 4.Creating folder to save graphs

```python
def Graph_Four(file,name,columns=[]):
    import pandas as pd
    import numpy as np
    import os
    import matplotlib.pyplot as plt
    import seaborn as sb
    file=pd.read_csv(file,index_col=0)
    
    
    if len(columns)==0:
        columns=list(file.columns)
 
    os.mkdir(name)
    os.chdir(name)
    numeric=[]
    categorical=[]
    
    
    for i in columns:
        if((np.array(file[i].unique().shape[0]>30) and file[i].dtype in ['float64','int64'])):
            numeric.append(i)
        else:
            categorical.append(i)
    
           
    for j in numeric:
        file.hist(column=j,grid=False,figsize=(8,6))
        plt.savefig(str(j)+'_hist.png')
        plt.show()   
           
           #boxplot
        file.boxplot(column=j,grid=False,figsize=(8,6))
        plt.savefig(str(j)+'_boxplot.png')
    for k in categorical:
        file[k].value_counts().plot(kind='bar',figsize=(6,4),color="blue",fontsize=13)
        plt.savefig(str(k)+'_barplot.png')
           
    os.chdir(r'C:\Users\Sulaiman\Downloads\datasets')       
```

```python
Graph_Four("cars.csv","graphs")
```
